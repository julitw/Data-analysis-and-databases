```python
#wyświetlenie oryginalej bazy
import pandas as pd
data = pd.read_csv('../OriginalData/tb.csv')
data
```

```python
# sprawdzenie czy w kolumnach iso2 i year występują NaN awartości - eliminacja wierszy
data_without_null = data.dropna(subset=['iso2', 'year'])

# usunięcie kolumn, gdzie nie ma przynajmniej 3 wartości różnych od Nan
data_without_null = data_without_null.dropna(axis=0, thresh=3)

# usunięcie kolumn dla przedziału wieku 0-4 i 5 -14 - ponieważ istnieje ogólna kolumna 0-14
data_without_null.drop(['new_sp_m04', 'new_sp_f04','new_sp_m514', 'new_sp_f514'], inplace=True, axis=1)

#zastąpienie wartości NaN zerami
df =  data_without_null.fillna(0)
df
```

```python
# zmiana formatu, gdzie kolumny 'iso2' i 'year'są zmiennymi identyfikatora a pozostałe to wartości mierzone
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[3:],
             var_name='column', value_name='cases')
df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[7]
df['age'] = df['column'].str[8:].map({
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+',
    'u': 'unknown'
})
df = df[['iso2', 'year', 'sex', 'age', 'cases']]

df.to_csv('../Analysisdata/tb-tidy.csv', index=False)   
df.to_csv('../Documents/tb-tidy.csv', index=False)
df = df[['iso2', 'year', 'sex', 'age', 'cases']]
df = df.sort_values(by=['iso2','year', 'age','sex'])
df = df.reset_index(drop=True)

# zapis wyczyszczonych danych do pliku
df.to_csv('../Analysisdata/tb-tidy.csv', index=False)   
df.to_csv('../Documents/tb-tidy.csv', index=False)
df
```
